```python
from scipy.stats import skew
import numpy as np
import pylab as p
x1=np.linspace(-5,5,1000)                             #spacing 5000 values from -5 to 5 linearly
y1=1./np.sqrt(2.*np.pi)*(np.exp(-.5*(x1**2)))         #formula for standard nor dist, Gaussian Distribution
p.plot(x1,y1,'*')
print('Skewness:', skew(y1))
```

```python
from scipy.stats import kurtosis
import numpy as np
import pylab as p
x1=np.linspace(-5,5,1000)
y1=1./np.sqrt(2.*np.pi)*(np.exp(-.5*(x1**2)))
p.plot(x1,y1,'*')
print('Kurtosis:', kurtosis(y1))
```
